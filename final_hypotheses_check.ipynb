# 2. Hypothesis

## In der Entertainment-Branche wird positiver kommentiert als in den anderen Branchen

```python
# Function to classify sentiment based on compound score
def classify_sentiment(compound):
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
# Apply the function to create a new column
us_comments['sentiment'] = us_comments['compound'].apply(classify_sentiment)
```

```python
us_comments.head()
```

```python
# Apply the function to create a new column
uk_comments['sentiment'] = uk_comments['compound'].apply(classify_sentiment)
```

```python
uk_comments.head()
```

Check whether the function was applied correct

```python
print("US Complete with detailed inspection:")
print(us_comments[['compound', 'sentiment']])
```

```python
print("UK Complete with detailed inspection:")
print(uk_comments[['compound', 'sentiment']])
```

```python
# Calculate percentage distribution of sentiment categories within each category
percentage_sentiment_distribution_uk = uk_comments.groupby(['category_name', 'sentiment']).size() / uk_comments.groupby('category_name').size() * 100

# Plot the percentage distribution
plt.figure(figsize=(10, 6))
percentage_sentiment_distribution_uk.unstack().plot(kind='bar', stacked=True, color=['orangered', 'lightgray', 'limegreen'])
plt.title('Percentage Distribution of Sentiment in the UK by Category')
plt.xlabel('Category')
plt.ylabel('Percentage of Comments')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

```python
percentage_sentiment_distribution_uk.loc[:, 'positive'].sort_values(ascending=False)
```

```python
# Calculate percentage distribution of sentiment categories within each category
percentage_sentiment_distribution_us = us_comments.groupby(['category_name', 'sentiment']).size() / us_comments.groupby('category_name').size() * 100

# Plot the percentage distribution
plt.figure(figsize=(10, 6))
percentage_sentiment_distribution_us.unstack().plot(kind='bar', stacked=True, color=['orangered', 'lightgray', 'limegreen'])
plt.title('Percentage Distribution of Sentiment in the US by Category')
plt.xlabel('Category')
plt.ylabel('Percentage of Comments')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

```python
percentage_sentiment_distribution_us.loc[:, 'positive'].sort_values(ascending=False)
```

The hypthesis that there is more positivity in the comment section of the entertainment branche is false. In the UK the Howto & Style Branche has the highest percentage of positive comments. In the US it's the Nonprofits & Activism category. In the UK the Entertainment branche is on 7th place and in the US only on 8th place.

# 1. Hypothese
## In den USA werden mehr Hate-Kommentare verbreitet als in GB

```python
uk_comments.value_counts('sentiment')
```

```python
number_negative_comments_uk = len(uk_comments[uk_comments['sentiment'] == 'negative'])
number_all_comments_uk = len(uk_comments)
```

```python
# percentage of negative comments in the UK
hate_percentage_uk = number_negative_comments_uk / number_all_comments_uk
print(f"Percentage of negative comments in the UK: {hate_percentage_uk * 100 :.2f}%")
```

and for the US

```python
number_negative_comments_us = len(us_comments[us_comments['sentiment'] == 'negative'])
number_all_comments_us = len(us_comments)
```

```python
# percentage of negative comments in the US
hate_percentage_us = number_negative_comments_us / number_all_comments_us
print(f"Percentage of negative comments in the US: {hate_percentage_us * 100 :.2f}%")
```

The hypothesis that there is more hate in the comment section of videos in the US is true, because there are 1.71% more hate comments in the US. It needs to be considered that this little difference is not significant, so one can say that the distribution of negative comments is the same in the UK and US.
