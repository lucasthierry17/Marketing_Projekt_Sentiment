```python
# Funktion zum Lesen einer JSON-Datei und Umwandeln in ein DataFrame
def json_to_dataframe(file):
    with open(file, 'r', encoding='utf-8') as json_file:
        data = json.load(json_file)
    
    # Extracting relevant information from JSON
    categories = []
    for item in data['items']:
        category_id = item['id']
        title = item['snippet']['title']
        categories.append({'category_id': category_id, 'title': title})
    
    # Creating DataFrame
    df = pd.DataFrame(categories)
    return df

# Beispielaufruf für den Gebrauch
# Ersetze 'file.json' durch den Pfad zu deiner eigenen Datei

# JSON-Datei in DataFrame umwandeln
gb_categories = json_to_dataframe('C:\\Users\\lucas\\OneDrive\\Documents\\Uni\\Sommersemester2024\\Marketing\\archive\\GB_category_id.json')
```
